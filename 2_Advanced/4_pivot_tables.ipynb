```python
from datasets import load_dataset
import matplotlib.pyplot as plt
import pandas as pd
my_dataset=load_dataset('lukebarousse/data_jobs')
df=my_dataset['train'].to_pandas()
df
```

```python
##Count of Job Titles

df.pivot_table(values='salary_year_avg', index='job_title_short',columns='job_country', aggfunc='median')
```

```python
top_country=df['job_country'].value_counts().head(6).index
top_country
df_top_country_salary=df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
df_top_country_salary=df_top_country_salary.loc[top_country]
df_top_country_salary
job_titles=['Data Analyst', 'Data Engineer', 'Data Scientist']
df_top_country_salary[job_titles].plot(kind='bar')
plt.title('Median Salary by Country and Job Title')
plt.ylabel('Median Salary in USD($)')
plt.xlabel('Country')
plt.xticks(rotation=45, ha='right')
```
