```python
import ast
from datasets import load_dataset
import matplotlib.pyplot as plt
import pandas as pd
```

```python
my_dataset=load_dataset('lukebarousse/data_jobs')
```

```python
df=my_dataset['train'].to_pandas()
```

```python
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
```

```python
df['job_skills']=df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df_explode=df.explode('job_skills')
```

```python
df_DA_Analyst=df_explode[df_explode['job_title_short']=='Data Analyst']
```

```python
df_DA_Analyst['job_posted_month_no']=df_DA_Analyst['job_posted_date'].dt.month
```

```python
df_DA_Analyst_pt=df_DA_Analyst.pivot_table(index='job_posted_month_no', columns='job_skills', aggfunc='size', fill_value=0)
```

```python
df_DA_Analyst_pt
```

```python
df_DA_Analyst_pt.loc['Total']=df_DA_Analyst_pt.sum()
```

```python
df_DA_Analyst_pt
```

```python
df_DA_Analyst_pt=df_DA_Analyst_pt[df_DA_Analyst_pt.loc['Total'].sort_values(ascending=False).index]
```

```python
df_DA_Analyst_pt
```

```python
df_DA_Analyst_pt=df_DA_Analyst_pt.drop('Total')
```

```python
df_DA_Analyst_pt=df_DA_Analyst_pt.reset_index()
```

```python
df_DA_Analyst_pt
```

```python
df_DA_Analyst_pt['job_posted_month']=df_DA_Analyst_pt['job_posted_month_no'].apply(lambda x: pd.to_datetime(str(x), format='%m').strftime('%B'))
```

```python
df_DA_Analyst_pt
```

```python
df_DA_Analyst_pt=df_DA_Analyst_pt.set_index('job_posted_month')
```

```python
df_DA_Analyst_pt
```

```python
df_DA_Analyst_pt=df_DA_Analyst_pt.drop(columns='job_posted_month_no')
```

```python
df_DA_Analyst_pt
```

```python
df_DA_Analyst_pt.iloc[:, :5].plot(
    kind='line',
    linewidth=4,
    linestyle='--',
    colormap='viridis',
    marker='o',
    markersize=5,
    figsize=(8, 8)
)
plt.xlabel=''
plt.ylabel='Count'
plt.show()
```
